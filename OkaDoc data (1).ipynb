```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
import pandas as pd
import time
import os

for j in range(2,11):
    

# driver.get method() will navigate to a page given by the URL address
#driver.get("https://www.okadoc.com/en-ae/doctor/family-medicine-and-general-medicine-gp")
    driver.get("https://www.okadoc.com/en-ae/doctor/family-medicine-and-general-medicine-gp?lang=en&country=ae&specialityTitle=doctor&specialityurl=family-medicine-and-general-medicine-gp&pageType=search-title-speciality-result&page={}".format(j))


    names = driver.find_elements_by_class_name("search-page__card__doctor__information__name") 
    locations = driver.find_elements_by_class_name("search-page__card__clinic__information__name") 

    nameList=[]
    locationList=[]

    for i in range(len(names)):
        name = names[i].text
        nameList.append(name)

    for i in range(len(locations)):
        location=locations[i].text
        locationList.append(location)


    #next=driver.find_element_by_xpath("//*[@id="id-search-page__card__section"]/div[2]/div[1]/ul/li[11]/a")
    #next.click()

    data_tuples = list(zip(nameList[0:],locationList[0:])) # list of each doctor's name and location paired together
    df = pd.DataFrame(data_tuples, columns=['Name','Location'])
    print(df)
    
    df.to_csv('Okadoc_Data.csv', mode='a', header=False)
   
```
