```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

driver = webdriver.Chrome(ChromeDriverManager().install())
import pandas as pd
import time
import os

for j in range(2,13):
    url="https://www.topdoctors.com.sa/en/doctors/riyadh/riyadh/?page={}".format(j)
    driver.get(url)

    for i in range(1,7):
        names = driver.find_elements(By.XPATH,'//*[@id="provider_result"]/ul/li["+str(i)+"]/article/div[1]/div/div[2]/div[2]/a/span')

    for i in range(1,7):
        specs = driver.find_elements(By.XPATH,'//*[@id="provider_result"]/ul/li["+str(i)+"]/article/div[1]/div/div[2]/div[2]/div/span')

    for i in range(1,7):
        clinics=driver.find_elements(By.XPATH,'//*[@id="provider_result"]/ul/li["+str(i)+"]/article/div[2]/div/div[1]/div/div[2]/button/span/span[2]/span/span[1]')

    nameList=[]
    specsList=[]
    clinicsList=[]


    for i in range(len(names)):
        name = names[i].text
        nameList.append(name)

    for i in range(len(specs)):
        spec=specs[i].text
        specsList.append(spec)

    for i in range(len(clinics)):
        clinic= clinics[i].text
        clinicsList.append(clinic)



    images = driver.find_elements_by_tag_name('img')
    imglist=[]
    for image in images:
        imgs=image.get_attribute('src')
        imglist.append(imgs)




    data_tuples = list(zip(nameList[0:],specsList[0:],clinicsList[0:],imglist[0:]))
    df = pd.DataFrame(data_tuples, columns=['Name','Specialization','Clinic','Photo'])

    print(df)
    print(j)
    df.to_csv('Top_Doctors.csv', mode='a', header=False)
```
